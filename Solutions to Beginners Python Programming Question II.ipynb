```python
import collections                                                                                      #importing function
import math                                                                                             #importing function
    
def d_func(fdata):                                                                                      #function defined for processing file
    fhand = fdata                                                                                       #storing filedata in fhand
    usr_dict = collections.defaultdict(dict)                                                            #creating the dictionary

    for line in fhand:                                                                                  #iterating over each line in fhand
        if len(line.strip())>0:                                                                         #condition check begins
            reviewer_name, reviews = line.strip('\n').lower().split(",",1)         #NOTE: **applying lower method here and the reason for that has been already explanined during our call** stripping, splitting and converitng the line to lowercase
            rating = reviews.split(",")                                                                 #splitting the restaurants and rating from data
            for rest in range(0,len(rating),2):                                                         #iterating over the restaurant
                usr_dict[reviewer_name][rating[rest]] = float(rating[rest+1])                           #storing the formatted data in dictionary

    rname_list = list(usr_dict.keys())                                                                  #creating list of dictionary keys
    rev_list = list(usr_dict.values())                                                                  #creating list of dictionary values
    
    
    def calculation(rev,reviewer):                                                                      #function defined to calculate the euclidean distance
        rest = set(usr_dict[rev].keys()).intersection(set(usr_dict[reviewer].keys()))                   #fetch the common restaurant among two user
        euc_dis = 0                                                                                     #initializing with zero
        for dis_btw in rest:                                                                            #iterating over common restaurants
            euc_dis += (usr_dict[rev][dis_btw]-usr_dict[reviewer][dis_btw])**2                          #calculating the difference, squaring them and sum it
        return math.sqrt(euc_dis)                                                                       #taking the square root of the distance and return the euclidean distance

    def one_to_many(rev):                                                                               #function defined to compute the review provided by one reviewer with respect to all other reviewer
        print("The Similarity Scores are:")                                                             #print statement
        for reviewer in usr_dict.keys():                                                                #iterating over each reviewer in dictionary
            if reviewer == rev:                                                                         #condition check if reviewer name is same as name provided by user
                continue                                                                                
            print(rev,reviewer,"{:.2f}".format((calculation(rev,reviewer))))                            #print the reviewer name, another reviewer names and their similarity score formatted upto 2 decimal place
    
    while True:                                                                                         #while loop begins

        usr_resp = input('\nWhat do you want to do? Input 1 for similarity between two reviewers, or Input 2 for similarity between one reviewer and all others in the database or 3 to quit: ').strip()                   #taking user choice as input
        if usr_resp == '1':                                                                             #condition check with user response
            rev_1 = input("Provide Reviewer1 name: ").lower().strip()                                   #ask user to input reviewer1 name and convert it to lowercase and strip
            rev_2 = input("Provide Reviewer2 name: ").lower().strip()                                   #ask user to input reviewer2 name and convert it to lowercase and strip
            if rev_1 in usr_dict.keys() and rev_2 in usr_dict.keys():                                   #condition check for reviewer name
                print('The similarity score between',rev_1 + ' and '+ rev_2 +' is:',"{:.4f}".format((calculation(rev_1,rev_2))))           #prints the similarity scores between two reviewer by calculation and formatting it
            elif rev_1 != usr_dict.keys() or rev_2 != usr_dict.keys():                                  #condition if reviewer name not present
                print('Please enter valid Reviewer name.')                                              #print statement when reviewer name not found
        
        elif usr_resp == '2':                                                                           #condition check with user response
            rev = input("Provide Reviewer name: ").lower().strip()                                      #ask user to input reviewer name and convert it to lowercase and strip 
            if rev in usr_dict.keys():                                                                  #check if reviewer name present in dictionary keys
                one_to_many(rev)                                                                        #calling function
            elif rev != usr_dict.keys():                                                                #condition check if reviewer name not in dictionary keys
                print('Please enter valid Reviewer name.')                                              #print this statement
        
        elif usr_resp == '3':                                                                           #condition check with user response
            print('Goodbye!')                                                                           #print statement
            break                                                                                       #break the flow
        
        elif usr_resp != '1' or usr_resp != '2' or usr_resp != '3':                                     #condition check if user provided response is other than 1, 2 or 3
            print('Invalid input, Please enter either 1, 2 or 3')                                       #print the statement 
       
    
try:                                                                                                    #try block starts
    usr_file = input("Give the name of the restaurant reviews file: ").lower().strip()                  #recieving filename as user input and converting to lowercase     
    fdata = open(usr_file,'r')                                                                          #opening the file in read mode
    d_func(fdata)                                                                                       #calling function
    quit()
except:                                                                                                 #except block
    print(usr_file,'file cannot be opened. Invalid filename or file not present.')                      #print statement when error is encountered
```
